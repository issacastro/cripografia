```python
import operator
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
def char_frequency(text):
    letras = {'a':0,'b':0,'c':0,'d':0,'e':0,'f':0,'g':0,'h':0,'i':0,'j':0,'k':0,'l':0,'m':0,'n':0,
              '単':0,'o':0,'p':0,'k':0,'r':0,'s':0,'t':0,'u':0,'v':0,'w':0,'x':0,'y':0,'z':0}
    for caracter in text:
        if caracter in letras.keys():
            letras[caracter] = letras[caracter] + 1
    return letras
```

```python
def Decifrar(texto,diccionario):
    txt=''
    for i in range(len(texto)):
        if texto[i] in diccionario.keys():
            txt+=diccionario[texto[i]]
        else:
            txt+=texto[i]
    return txt
```

```python
Frec =pd.read_csv('Frecuencias.csv',index_col=0)
Frec.sort_values(by='Frec',ascending = False)
```

```python
texto =open("texto_m.txt",'r',encoding="utf8").read().lower()
texto = texto.replace(' ','')
F= pd.DataFrame.from_dict(char_frequency(texto),orient='index',columns=['Frec'])
F.to_dict('dict')
F['Probabilidad'] = F['Frec']/F['Frec'].sum()*100
F.sort_values(by='Frec',ascending = False)
```

```python
data = pd.DataFrame(data=Frec.sort_values(by='Frec',ascending = False).index,index=F.sort_values(by='Frec',ascending = False).index,columns=['D'])
DECRYPT = data.to_dict('dict')
#texto =open("Isma-Cifrada.txt",'r',encoding="utf8").read().lower()
texto =open("texto_m.txt",'r',encoding="utf8").read().lower()
x=Decifrar(texto,DECRYPT['D'])
print(x)
filename='Carta-Decifrada.txt'
filew = open(filename, "w",encoding="utf-8")
filew.write(x)
filew.close()
```
