```python
import random

"""
DES Key Scheduler:

    - Generates 16 48-bit subkeys from a randomly
      generated 64-bit initial key

    - Returns a list of the 16 subkeys in the form
      of binary strings

"""

def key_scheduler(k):

    # -------------------- Variables -------------------- #
     # k Clave inicial de 64 bits
     # k_prime Clave permutada de 56 bits
     # c0 Mitad izquierda de k_prime
     # d0 Mitad derecha de k_prime
     # c_keys 16 permutaciones de 28 bits de c0 usando lrt
     # d_keys 16 permutaciones de 28 bits de d0 usando lrt
     # claves Lista final de 16 claves de 48 bits
    #
    # --------------------------------------------------- #

    
    # ---------------- Hard-Coded Values ---------------- #

    # TABLA pc1      
    pc1 = [57, 49, 41, 33, 25, 17,  9,
            1, 58, 50, 42, 34, 26, 18,
           10,  2, 59, 51, 43, 35, 27,
           19, 11,  3, 60, 52, 44, 36,
           63, 55, 47, 39, 31, 23, 15,
            7, 62, 54, 46, 38, 30, 22,
           14,  6, 61, 53, 45, 37, 29,
           21, 13,  5, 28, 20, 12,  4]

    # TABLA pc2       
    pc2 = [14, 17, 11, 24,  1,  5,
           3,  28, 15,  6, 21, 10,
           23, 19, 12,  4, 26,  8,
           16,  7, 27, 20, 13,  2,
           41, 52, 31, 37, 47, 55,
           30, 40, 51, 45, 33, 48,
           44, 49, 39, 56, 34, 53,
           46, 42, 50, 36, 29, 32]
           

    # TABA DE ROTACION
    lrt = [1, 1, 2, 2, 2, 2, 2, 2,
           1, 2, 2, 2, 2, 2, 2, 1]

    # --------------------------------------------------- #


    
# Generar un número aleatorio de 64 bits
     # Nota: aunque el algoritmo comienza con una clave de 64 bits, DES todavía es técnicamente
     # 56 bits, porque 8 de los bits se utilizan como bits de paridad. Estos bits se pierden durante el
     # primera permutación con pc1.
   # k = 9024365068935622739
    #k = bin(k)[2:].zfill(64)
    

    # Generar k_prime usando pc1
    k_prime = ""
    for number in pc1:
        k_prime += k[number-1]
        

    #Encuentre c0 (mitad izquierda de k_prime) y d0 (mitad derecha de k_prime)
    c0 = k_prime[:28]
    d0 = k_prime[28:]


    # Rotar c0 y d0 de acuerdo con la tabla de rotación izquierda
    c_keys = []
    d_keys = []
    for i in range(16):
        num = lrt[i]
        
        temp = c0[0:num]
        c0 = c0[num:] + temp
        c_keys.append(c0)

        temp = d0[0:num]
        d0 = d0[num:] + temp
        d_keys.append(d0)



# Generar subclaves combinando c_keys [i] y d_keys [i].
# Use la tabla pc2 para generar 16 claves de 48 bits
    keys = []
    for i in range(16):
        cat = c_keys[i] + d_keys[i]
        temp = ""
        for number in pc2:
            temp += cat[number-1]
        keys.append(temp)


    return keys


```
