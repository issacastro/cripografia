```python
dicc =  {'A':1,'B':2,'C':3,'D':4,'E':5,'F':6,'G':7,
         'H':8,'I':9,'J':10,'K':11,'L':12,'M':13,'N':14,
         'Ñ':15,'O':16,'P':17,'Q':18,'R':19,'S':20,'T':21,
         'U':22,'V':23,'W':24,'X':25,'Y':26,'Z':27,' ':28}
#Convierte el texto en m de longitud de 10 
m_len = 4
def text2m(texto):
    x =[]
    m=''
    for i in range(len(texto)):
        if dicc[texto[i]] <10:
            m += '0'+ str(dicc[texto[i]])
        else:
            m +=str(dicc[texto[i]])
        if len(m) == m_len:
            x.append(m)
            m=''
        if len(m) <m_len and i == len(texto)-1:
            while(len(m)<m_len):
                m+='0'
            x.append(m)
    encrypted_blocks =[]
    for i in range(len(x)):
        encrypted_blocks.append(int(x[i]))
        
    return encrypted_blocks  

def m2text(x):
    text =''
    for num in x:
        m= str(num)
        if len(m)<m_len:
            m = '0' + m
        for i in range (0,len(m),2):
            if int(m[i:2+i]) != 0:
                text+=list(dicc.keys())[list(dicc.values()).index(int(m[i:2+i]))]
    return text
```

```python
"""
Este programa implementa el algoritmo RSA para la criptografía.
Selecciona aleatoriamente dos números primos de un archivo txt de números primos y
los usa para producir las claves públicas y privadas.
"""
import random

def gcd(a, b):
    """
    Realiza el algoritmo euclidiano y devuelve el mcd de a y b
    """
    if (b == 0):
        return a
    else:
        return gcd(b, a % b)

def xgcd(a, b):
    """
    Realiza el algoritmo euclidiano extendido.Devuelve el mcd, el coeficiente de a y el coeficiente de b
    """
    x, old_x = 0, 1
    y, old_y = 1, 0

    while (b != 0):
        quotient = a // b
        a, b = b, a - quotient * b
        old_x, x = x, old_x - quotient * x
        old_y, y = y, old_y - quotient * y

    return a, old_x, old_y

def chooseE(fi):
    """
    Elige un número aleatorio, 1 < e <fi, y comprueba si es o no
    coprime con el fi, es decir, mcd (e, fi) = 1
    """
    while (True):
        e = random.randrange(2, fi)

        if (gcd(e, fi) == 1):
            return e

def chooseKeys(aleatorio):
    """
    Selecciona dos números primos aleatorios de una lista de números primos que tiene
    valores que van hasta 100k. Crea un archivo de texto y almacena los dos
    números allí donde se pueden usar más tarde. Usando los números primos,
    También calcula y almacena las claves públicas y privadas en dos
    archivos
    """
    if aleatorio =='SI':
        """
        Esta parte es si quieres elegir dos numeros primos aleatorio
        """
        #elije dos números aleatorios dentro del rango de líneas donde
        # los números primos no son demasiado pequeños ni demasiado grandes
        rand1 = random.randint(100, 300)
        rand2 = random.randint(100, 300)

        #  lee el archivo txt de números primos en una lista de python
        fo = open('primes-to-100k.txt', 'r')
        lines = fo.read().splitlines()
        fo.close()

        # almacenar nuestros números primos en estas variables
        p = int(lines[rand1])
        q = int(lines[rand2])
        print('p:',p,'q:',q)

    else:
        """
        Esta parte es si quieres elegir dos numeros primos aleatorio
        """
        p = int(input('Ingrese p:')) #838367
        q = int(input('Ingrese q:')) #838141
    # calular n, fi, e
    n = p * q
    fi = (p - 1) * (q - 1)
    e = chooseE(fi)

    # calcular d, 1 < d < fi tal que ed = 1 (mod fi)
    # e y d son inversos (mod fi)
    gcd, x, y = xgcd(e, fi)

    # se asegura que sea positivo
    if (x < 0):
        d = x + fi
    else:
        d = x

    #guardar las claves públicas n y e en un archivo
    f_public = open('public_keys.txt', 'w')
    f_public.write(str(n) + '\n')
    f_public.write(str(e) + '\n')
    f_public.close()

    f_private = open('private_keys.txt', 'w')
    f_private.write(str(n) + '\n')
    f_private.write(str(d) + '\n')
    f_private.close()
    ed=e*d
    mod = 1 % fi
    print('fi:',fi,'e:',e,'d:',d,'ed=',ed,"1 mod fi=",mod)

def encrypt(message,key):
 
    n = int(key.readline())
    e = int(key.readline())
    key.close()

    encrypted_blocks = text2m(message)
    #print(encrypted_blocks)
    x =''
    for i in range(len(encrypted_blocks)):
        encrypted_blocks[i] = str((pow(encrypted_blocks[i],e))% n) 


    # crear una cadena a partir de los números
    encrypted_message = " ".join(encrypted_blocks)
    return encrypted_message

def decrypt(blocks,key):
    """
    Descifra una cadena de números elevando cada número a la potencia de d y
    tomando el módulo de n. Devuelve el mensaje como una cadena.
    block_size se refiere a cuántos caracteres forman un grupo de números en
    Cada índice de bloques.
    """

    #fo = open('public_keys.txt', 'r')
    n = int(key.readline())
    d = int(key.readline())
    key.close()

    # convierte la cadena en una lista de entradas
    list_blocks = blocks.split(' ')
    int_blocks = []
    x =[]
    for s in list_blocks:
        int_blocks.append(int(s))
    #print(int_blocks)
    message = ""
    for i in range(len(int_blocks)):
        x.append((pow(int_blocks[i],d)) % n)
    #print(x)
    message = m2text(x)
    return message
```
